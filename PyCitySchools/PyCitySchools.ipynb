```python
import pandas as pd
```

```python
stupath = "Resources/students_complete.csv"
schlpath = "Resources/schools_complete.csv"

stulis_df = pd.read_csv(stupath)
school_df = pd.read_csv(schlpath)

#stulis_df.head()
#school_df.head(16)
```

```python
stubyschl_df = stulis_df.groupby('school_name').mean()
read_scr_schl = [round(rval,2) for rval in stubyschl_df['reading_score']]
math_scr_schl = [round(mval,2) for mval in stubyschl_df['math_score']]
stubyschl_df.head()

#these lists should be sorted by school in alphabetical order
#print(read_scr_schl)
#print(math_scr_schl)
```

```python
stubypass_read_df = stulis_df.loc[(stulis_df['reading_score'] >= 70)]
stubypass_math_df = stulis_df.loc[(stulis_df['math_score'] >= 70)]
stubypass_df = stulis_df.loc[(stulis_df['reading_score'] >= 70) & (stulis_df['math_score'] >= 70)]

stubypass_df = stubypass_df.groupby('school_name').count()
stubypass_read_df = stubypass_read_df.groupby('school_name').count()
stubypass_math_df = stubypass_math_df.groupby('school_name').count()
stubypass = stubypass_df['student_name'].tolist()
stubypass_read = stubypass_read_df['reading_score'].tolist()
stubypass_math = stubypass_math_df['math_score'].tolist()

#these lists should be sorted by school in alphabetical order and contain the numberr of students that meet passing criteria 
#print(stubypass)
#print(stubypass_read)
#print(stubypass_math)
```

```python
#District Summary
#Create Data Frame to include Total Number of unique schools; total number of students; total budget; average math score; average reading score; % passing math, % passing reading, % passing both

school_count = school_df.school_name.unique().size
#print("There are " + str(school_count) + " schools in this distric.")
student_count = school_df['size'].sum()
#print("There are " + str(student_count) + " students in this distric.")
total_budget = school_df['budget'].sum()
#print("The total budget for all schools in the distric is $" + str(total_budget) + ".")
read_score_total = stulis_df['reading_score'].sum()
dis_avr_read = round((read_score_total)/(student_count),2)
#print("The average grade in reading in the distric is " + str(dis_avr_read) + ".")
math_score_total = stulis_df['math_score'].sum()
dis_avr_math = round((math_score_total)/(student_count),2)
#print("The average grade in math in the distric is " + str(dis_avr_math) + ".")
num_pass_read = len(stulis_df[stulis_df['reading_score']>=70])
prcnt_pass_read = round(((num_pass_read)/(student_count)*100),2)
#print("The percent of students passing reading in the distric is " + str(prcnt_pass_read) + "%.")
num_pass_math = len(stulis_df[stulis_df['math_score']>=70])
prcnt_pass_math = round(((num_pass_math)/(student_count)*100),2)
#print("The percent of students passing math in the distric is " + str(prcnt_pass_math) + "%.")
num_pass_both = len(stulis_df[(stulis_df['math_score']>=70) & (stulis_df['reading_score']>=70)])
prcnt_pass_both = round(((num_pass_both)/(student_count)*100),2)
#print("The percent of students passing both reading and math in the distric is " + str(prcnt_pass_both) + "%.")
```

```python
distric_summary = {
    'Number of Schools':[school_count],
    'Number of Students':[student_count],
    'Total Budget':[total_budget],
    'Average Math Score':[dis_avr_math],
    'Average Reading Score':[dis_avr_read],
    'Percent Passing Math':[prcnt_pass_math],
    'Percent Passing Reading':[prcnt_pass_read],
    'Percent Passing Both':[prcnt_pass_both]
    }
distric_summary_df = pd.DataFrame(distric_summary)
#distric_summary_df.style.hide(axis='index')
```

```python
#School Summary
#Create Data Frame to include School name, school type, total number of students; total budget; budget per student, average math score; average reading score; % passing math, % passing reading, % passing both
#build new dataframe off of school_df
#need to group stulis_df by school and then later further by grade
#add additional columns to new df based on calculations done on stulis_df groups

#sort dataframe to be alphabetical 
school_df = school_df.sort_values(by=['school_name'])
del school_df['School ID']
school_df = school_df.rename(columns={"school_name":"School Name","type":"School Type","size":"Number of Students","budget":"Budget"})
school_df['Average Reading Score'] = read_scr_schl
school_df['Average Math Score'] = math_scr_schl
school_df['Number of Students Passing Reading'] = stubypass_read
school_df['Number of Students Passing Math'] = stubypass_math
school_df['Number of Students Passing Overall'] = stubypass
#school_df
```

```python
for p in school_df:
    school_df["Percent of Students Passing Reading"] = round((((school_df["Number of Students Passing Reading"])/(school_df["Number of Students"]))*100),2)
    school_df["Percent of Students Passing Math"] = round((((school_df["Number of Students Passing Math"])/(school_df["Number of Students"]))*100),2)
    school_df["Percent of Students Passing Overall"] = round((((school_df["Number of Students Passing Overall"])/(school_df["Number of Students"]))*100),2)
    school_df["Budget per Student"] = round((((school_df["Budget"])/(school_df["Number of Students"]))),2)
    
#school_df
```

```python
del school_df['Number of Students Passing Reading']
del school_df['Number of Students Passing Math']
del school_df['Number of Students Passing Overall']
```

```python
school_df = school_df[['School Name', 'School Type', 'Number of Students', 'Budget', 'Budget per Student', 'Average Reading Score', 'Average Math Score', 'Percent of Students Passing Reading', 'Percent of Students Passing Math', 'Percent of Students Passing Overall']]
school_df.reset_index(inplace=True, drop=True)

#school_df.style.hide(axis='index')
```

```python
top_schools = school_df.sort_values(by=['Percent of Students Passing Overall'], ascending=False)
#top_schools.head(5).style.hide(axis='index')
```

```python
bottom_schools = school_df.sort_values(by=['Percent of Students Passing Overall'], ascending=True)
#bottom_schools.head(5).style.hide(axis='index')
```

```python
stulis_df['grade'] = stulis_df['grade'].replace({"9th":9,"10th":10,"11th":11,"12th":12})
stulis_df['grade'].astype('int')
scrbygrade_df = stulis_df.groupby(['school_name','grade']).mean()

del scrbygrade_df['Student ID']
scrbygrade_df = scrbygrade_df.rename(columns={"school_name":"School Name","grade":"Grade Level","reading_score":"Average Reading Score","math_score":"Average Math Score"})

for o in scrbygrade_df:
    scrbygrade_df['Average Reading Score'] = round(scrbygrade_df['Average Reading Score'],2)
    scrbygrade_df['Average Math Score'] = round(scrbygrade_df['Average Math Score'],2)


#scrbygrade_df
```

```python
spending_school_df = school_df

spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
spending_school_df["Spending Range per Student"] = pd.cut(spending_school_df['Budget per Student'],spending_bins, labels=labels,include_lowest=True)

#spending_school_df
```

```python
spending_math_scores = spending_school_df.groupby(["Spending Range per Student"]).mean()["Average Math Score"]
spending_reading_scores = spending_school_df.groupby(["Spending Range per Student"]).mean()["Average Reading Score"]
spending_passing_math = spending_school_df.groupby(["Spending Range per Student"]).mean()["Percent of Students Passing Math"]
spending_passing_reading = spending_school_df.groupby(["Spending Range per Student"]).mean()["Percent of Students Passing Reading"]
spending_passing_overall = spending_school_df.groupby(["Spending Range per Student"]).mean()["Percent of Students Passing Overall"]
```

```python
spending_summary = pd.merge(spending_math_scores,spending_reading_scores, on="Spending Range per Student")
spending_summary = pd.merge(spending_summary, spending_passing_math, on="Spending Range per Student")
spending_summary = pd.merge(spending_summary, spending_passing_reading, on="Spending Range per Student")
spending_summary = pd.merge(spending_summary, spending_passing_overall, on="Spending Range per Student")

for p in spending_summary:
    spending_summary['Average Reading Score'] = round(spending_summary['Average Reading Score'],2)
    spending_summary['Average Math Score'] = round(spending_summary['Average Math Score'],2)
    spending_summary['Percent of Students Passing Math'] = round(spending_summary['Percent of Students Passing Math'],2)
    spending_summary['Percent of Students Passing Reading'] = round(spending_summary['Percent of Students Passing Reading'],2)
    spending_summary['Percent of Students Passing Overall'] = round(spending_summary['Percent of Students Passing Overall'],2)

#spending_summary
```

```python
sizeof_school_df = school_df

size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
spending_school_df["Size of School"] = pd.cut(spending_school_df['Number of Students'],size_bins, labels=labels,include_lowest=True)

#sizeof_school_df
```

```python
size_math_scores = sizeof_school_df.groupby(["Size of School"]).mean()["Average Math Score"]
size_reading_scores = sizeof_school_df.groupby(["Size of School"]).mean()["Average Reading Score"]
size_passing_math = sizeof_school_df.groupby(["Size of School"]).mean()["Percent of Students Passing Math"]
size_passing_reading = sizeof_school_df.groupby(["Size of School"]).mean()["Percent of Students Passing Reading"]
size_passing_overall = sizeof_school_df.groupby(["Size of School"]).mean()["Percent of Students Passing Overall"]
```

```python
size_summary = pd.merge(size_math_scores,size_reading_scores, on="Size of School")
size_summary = pd.merge(size_summary, size_passing_math, on="Size of School")
size_summary = pd.merge(size_summary, size_passing_reading, on="Size of School")
size_summary = pd.merge(size_summary, size_passing_overall, on="Size of School")

for q in size_summary:
    size_summary['Average Reading Score'] = round(size_summary['Average Reading Score'],2)
    size_summary['Average Math Score'] = round(size_summary['Average Math Score'],2)
    size_summary['Percent of Students Passing Math'] = round(size_summary['Percent of Students Passing Math'],2)
    size_summary['Percent of Students Passing Reading'] = round(size_summary['Percent of Students Passing Reading'],2)
    size_summary['Percent of Students Passing Overall'] = round(size_summary['Percent of Students Passing Overall'],2)

#size_summary
```

```python
typeof_school_df = school_df

type_math_scores = typeof_school_df.groupby(["School Type"]).mean()["Average Math Score"]
type_reading_scores = typeof_school_df.groupby(["School Type"]).mean()["Average Reading Score"]
type_passing_math = typeof_school_df.groupby(["School Type"]).mean()["Percent of Students Passing Math"]
type_passing_reading = typeof_school_df.groupby(["School Type"]).mean()["Percent of Students Passing Reading"]
type_passing_overall = typeof_school_df.groupby(["School Type"]).mean()["Percent of Students Passing Overall"]
```

```python
type_summary = pd.merge(type_math_scores,type_reading_scores, on="School Type")
type_summary = pd.merge(type_summary, type_passing_math, on="School Type")
type_summary = pd.merge(type_summary, type_passing_reading, on="School Type")
type_summary = pd.merge(type_summary, type_passing_overall, on="School Type")

for b in type_summary:
    type_summary['Average Reading Score'] = round(type_summary['Average Reading Score'],2)
    type_summary['Average Math Score'] = round(type_summary['Average Math Score'],2)
    type_summary['Percent of Students Passing Math'] = round(type_summary['Percent of Students Passing Math'],2)
    type_summary['Percent of Students Passing Reading'] = round(type_summary['Percent of Students Passing Reading'],2)
    type_summary['Percent of Students Passing Overall'] = round(type_summary['Percent of Students Passing Overall'],2)

#type_summary
```
